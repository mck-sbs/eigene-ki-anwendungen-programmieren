```python
import pandas as pd
import tensorflow as tf
from keras.models import Model
import numpy as np
```

```python
# Erzeuge Zufallszahlen 0-99
input_array = np.random.randint(100, size=(2, 4))
print(input_array)
```

```python
emb = tf.keras.layers.Embedding(100, 3, input_length=4)

model = tf.keras.Sequential()
model.add(emb)

output_array = model.predict(input_array)
print(output_array)

model.summary()
```

```python
emb = tf.keras.layers.Embedding(100, 3, input_length=4)

model = tf.keras.Sequential()
model.add(emb)

output_array = model.predict(input_array)
print(output_array)


model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.GlobalAveragePooling1D())

output_array2 = model2.predict(output_array)
print(output_array2)
```
