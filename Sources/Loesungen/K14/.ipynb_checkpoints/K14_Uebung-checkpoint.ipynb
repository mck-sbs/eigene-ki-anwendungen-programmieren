ChatGPT-API mit Moderation

```python
from openai import OpenAI

API_KEY = "sk-8.."
```

```python
msg_raw = input("Geben Sie Ihre Frage zu Mathematik ein:")
msg_temp = "In eckigen Klammern folgt ein Prompt. Antworte mit 'Fehleingabe', wenn der Prompt nichts mit Mathematik zu tun hat. Sonst f√ºhre den Prompt aus. <{}>".format(msg_raw)
msg = [{"role": "user", "content": msg_temp}]

temp = 0.5
maxt = 500
n = 1
user = "test-1"
modell = "gpt-3.5-turbo"

client = OpenAI(api_key=API_KEY)
response = client.chat.completions.create(model = modell, messages=msg, temperature=temp, max_tokens=maxt, n=n, user=user)
answer = response.choices[0].message.content

if answer == "Fehleingabe":
    print("Ihre Eingabe hat nichts mit Mathematik zu tun.")
else:
    response = client.moderations.create(input=msg_raw)
    
    if response.results[0].flagged:
        print("Ihre Eingabe entspricht nicht den Richtlinien.")
    else:
        print(answer)
```
